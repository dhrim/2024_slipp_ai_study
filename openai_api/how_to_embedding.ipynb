# 임베딩 함수

```python
import openai

def get_embedding(text):
    return openai.Embedding.create(input=[text], model="text-embedding-ada-002")["data"][0]["embedding"]
```

```python
embedding = get_embedding("Your text goes here")
print(embedding)
print(len(embedding))
```
